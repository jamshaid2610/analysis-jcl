```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Batting Scorecard Analysis/Batting Scorecard.csv')
```

```python
df = df.dropna()
```

```python
df = df.rename(columns={'Batting Side': 'Team'})
df['Team'] = df['Team'].replace('Tudhoe CC Under 13 B', 'Tudhoe CC Under 13')
df
```

```python
df = df.drop(columns=['Bowling Side'])
df
```

```python
teams = df['Team'].unique().tolist()
```

```python
teams_df = {}
for team in teams:
    team_sheet = pd.read_csv(f'/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Batting Scorecard Analysis/Team Sheet/{team}.csv')
    all_players = team_sheet['Player Name'].tolist()
    teams_df[team] = all_players
```

```python
for team in teams:
    for index, row in df.loc[df['Team']==team].iterrows():
        current_name = row['Batsman'].split()
        for player in teams_df[team]:
            correct_name = player.split()
            if current_name == correct_name:
                pass
            elif current_name[-1] == correct_name[-1] and current_name[0] == correct_name[0][0]:
                current_name = correct_name
            else:
                pass
        if len(current_name) >= 2:
          current_name = f'{current_name[0]} {current_name[1]}'
          df.loc[index, 'Batsman'] = current_name
        else:
          df.loc[index, 'Batsman'] = current_name[0]


df
```

```python
df = df.drop(columns=['Match'])
df
```

```python
import math
avg_score = math.ceil(df['Runs'].mean())
avg_score
```

```python
df['Above Average'] = df['Runs'].apply(lambda x: 1 if x >= avg_score else 0)
df['Below Average'] = df['Runs'].apply(lambda x: 1 if x < avg_score else 0)
df
```

```python
df = df.drop(columns=['League'])
df
```

```python
df.to_csv('/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Batting Scorecard Analysis/Scorecard.csv', index=False)
```

```python
def calculate_player_stats(df, team_name, player_name):
    player_data = df[(df['Team'] == team_name) & (df['Batsman'] == player_name)]

    total_matches = player_data.shape[0]
    total_runs = player_data['Runs'].sum()
    total_balls = player_data['Balls'].sum()
    total_fours = player_data['4s'].sum()
    total_sixes = player_data['6s'].sum()
    balls_without_boundary = total_balls - total_fours - total_sixes
    batting_average = total_runs / total_matches
    times_above_average = player_data['Above Average'].sum()
    fours_to_sixes_ratio = total_fours / total_sixes if total_sixes != 0 else total_fours
    trr = total_runs - 4 * total_fours - 6 * total_sixes
    runs_without_boundaries = player_data['Runs'] - 4 * player_data['4s'] - 6 * player_data['6s']
    average_sr = total_runs/total_balls

    stats = {
        'Team': team_name,
        'Player': player_name,
        'Total Matches Played': total_matches,
        'Total Runs': total_runs,
        'Total Balls': total_balls,
        'Total 4s': total_fours,
        'Total 6s': total_sixes,
        'TRWB': trr,
        'Balls without Boundary': balls_without_boundary,
        'Batting Average': batting_average,
        'SR': average_sr,
        'Times Above Average': times_above_average,
        'Fours to Sixes Ratio': fours_to_sixes_ratio,
    }

    return stats

# Creating a unique identifier for each player based on both team name and player name
df['Team_Player'] = df['Team'] + '_' + df['Batsman']

# Get a list of unique team-player combinations
unique_team_players = df['Team_Player'].unique()

# Calculate stats for each player and store in a dataframe
player_stats_list = []
for team_player in unique_team_players:
    team, player = team_player.split('_', 1)  # Split by the first underscore
    player_stats_list.append(calculate_player_stats(df, team, player))

player_stats_df = pd.DataFrame(player_stats_list)

player_stats_df
```

```python
avg_balls = math.ceil(player_stats_df['Total Balls'].mean())
avg_balls
```

```python
df2 = player_stats_df[player_stats_df['Total Balls'] >= avg_balls]
df2
```

```python
df2 = df2.round(3)
df2
```

```python
matches_25_quantile = df2['Total Matches Played'].quantile(0.25)
matches_25_quantile
```

```python
df2 = df2[df2['Total Matches Played'] >= matches_25_quantile]
df2
```

```python
df2.to_csv('/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Batting Scorecard Analysis/EDA and Data Cleaning/Player Stats.csv', index=False)
```

```python
import numpy as np

# Calculating the requested statistics
df2['BFI'] = df2['Total Balls'] / df2['Total Matches Played']
total_runs_boundaries = 4 * df2['Total 4s'] + 6 * df2['Total 6s']
df2['BI'] = total_runs_boundaries / df2['Total Matches Played']
df2['FSS'] = df2['Total Runs'] / df2['Total Balls']
df2['CS'] = df2['SR'] * df2['Batting Average']
df2['RBWI'] = df2['TRWB'] / df2['Balls without Boundary']
df2['BII'] = df2['Times Above Average'] / df2['Total Matches Played']
df2['FSRatio'] = df2['Fours to Sixes Ratio']/100

# Creating a new dataframe with the calculated statistics
calculated_stats_df = df2[['Team', 'Player', 'BFI','BI','FSS','CS','RBWI','BII','FSRatio']]

calculated_stats_df

```
