```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Ball By Ball Analysis/Batting Analysis (Ball By Ball)/Player Statistics.csv')
```

```python
df
```

```python
df['BFI'] = df['Total Balls'] / df['Total Matches Played']
df['BI'] = df['Runs with Boundaries']/ df['Total Matches Played']
df['FSS'] = df['Total Runs'] / df['Total Balls']
df['CS'] = df['SR'] * df['Batting Average']
df['RBWI'] = df['TRWB'] / df['Balls without Boundary']
df['BII'] = df['Times Above Average'] / df['Total Matches Played']
df['FSRatio'] = df['Fours to Sixes Ratio']/100
df['DBI'] = df['Total Dot Balls'] / df['Total Matches Played']
df['FS'] = df['Total Not Dismissed'] / df['Total Matches Played']

df2 = df[['Team', 'Player', 'BFI', 'BI', 'FSS', 'CS', 'RBWI', 'BII', 'FSRatio', 'DBI', 'FS']]

df2
```

```python
df2 = df2.round(4)
df2
```
