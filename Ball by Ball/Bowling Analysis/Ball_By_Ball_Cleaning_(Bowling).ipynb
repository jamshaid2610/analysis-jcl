```python
def calculate_overs_decimal(balls_without_extras):
    whole_overs = balls_without_extras // 6
    remaining_balls = balls_without_extras % 6
    if remaining_balls > 6:
        whole_overs += 1
        remaining_balls -= 6
    return whole_overs + remaining_balls / 10

grouped_df = df.groupby(['Match', 'Team', 'Bowler']).agg(
    Balls_Bowled_With_Extras=('Bowler', 'count'),
    Extras=('Run Type', lambda x: (x == 'extras').sum()),
    Runs=('Runs', lambda x: pd.to_numeric(x, errors='coerce').sum()),
    Wickets=('Run Type', lambda x: (x == 'dismissed').sum()),
    Dot_Balls=('Run Type', lambda x: (x == 'dot ball').sum()),
    Boundaries=('Run Type', lambda x: (x == 'boundary').sum()),
)

grouped_df['Balls_Bowled_Without_Extras'] = grouped_df['Balls_Bowled_With_Extras'] - grouped_df['Extras']
grouped_df['Overs_Bowled'] = grouped_df['Balls_Bowled_Without_Extras'].apply(calculate_overs_decimal)
grouped_df['3W'] = (grouped_df['Wickets'] == 3).astype(int)
grouped_df['4W'] = (grouped_df['Wickets'] == 4).astype(int)
grouped_df['5W'] = (grouped_df['Wickets'] == 5).astype(int)


df2 = grouped_df.copy().reset_index()
df2
```

```python
df2 = df2.drop(columns=['Match']).reset_index(drop=True)
df2
```

```python
def calculate_player_stats(df, team_name, player_name):
  player_data = df[(df['Team']== team_name) & (df['Bowler'] == player_name)]

  total_matches = player_data.shape[0]
  total_runs = player_data['Runs'].sum()
  total_wickets = player_data['Wickets'].sum()
  total_balls_bowled = player_data['Balls_Bowled_With_Extras'].sum()
  total_ball_bowled_without_extras = player_data['Balls_Bowled_Without_Extras'].sum()
  total_overs = calculate_overs_decimal(total_ball_bowled_without_extras)
  total_extras = player_data['Extras'].sum()
  total_3Ws = player_data['3W'].sum()
  total_4Ws = player_data['4W'].sum()
  total_5Ws = player_data['5W'].sum()
  average_balls_bowled = total_balls_bowled / total_matches
  strike_rate = (total_wickets/total_ball_bowled_without_extras) if total_wickets > 0 else 0
  economy_rate = (total_runs/total_ball_bowled_without_extras)
  total_matches_with_3W_plus = total_3Ws + total_4Ws + total_5Ws
  total_wickets_3W_plus = 3*total_3Ws + 4*total_4Ws + 5*total_5Ws
  total_wickets_less_than_3W = total_wickets - total_wickets_3W_plus
  consistency_score = total_runs / total_wickets if total_wickets > 0 else 0
  total_dot_balls = player_data['Dot_Balls'].sum()
  total_boundaries = player_data['Boundaries'].sum()


  stats = {
      'Team': team_name,
      'Player': player_name,
      'Total Matches': total_matches,
      'Total Overs': total_overs,
      'Total Runs': total_runs,
      'Total Wickets': total_wickets,
      'Total Balls Bowled': total_balls_bowled,
      'Total Balls Bowled Without Extras': total_ball_bowled_without_extras,
      'Total Extras': total_extras,
      'Total 3Ws': total_3Ws,
      'Total 4Ws': total_4Ws,
      'Total 5Ws': total_5Ws,
      'Average Balls Bowled': average_balls_bowled,
      'Strike Rate': strike_rate,
      'Economy Rate': economy_rate,
      'Total Matches with 3W+': total_matches_with_3W_plus,
      'Total Wickets with 3W+': total_wickets_3W_plus,
      'Total Wickets less than 3W': total_wickets_less_than_3W,
      'Consistency Score': consistency_score,
      'Total Dot Balls': total_dot_balls,
      'Total Boundaries': total_boundaries,
  }

  return stats

df2['Team_Player'] = df2['Team'] + '_' + df2['Bowler']
unique_team_players = df2['Team_Player'].unique()

player_stats_list = []
for team_player in unique_team_players:
  team, player = team_player.split('_',1)
  player_stats_list.append(calculate_player_stats(df2, team, player))

player_stats_df = pd.DataFrame(player_stats_list)

player_stats_df
```
