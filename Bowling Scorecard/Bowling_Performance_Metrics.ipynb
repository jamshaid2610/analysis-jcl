```python
df = df.dropna()
```

```python
df = df.rename(columns={'Bowling Side': 'Team'})
df['Team'] = df['Team'].replace('Tudhoe CC Under 13 B', 'Tudhoe CC Under 13')
df
```

```python
df = df.drop(columns=['Batting Side'], axis=1)
df
```

```python
teams = df['Team'].unique().tolist()
```

```python
teams_df = {}
for team in teams:
    team_sheet = pd.read_csv(f'/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Bowling Scorecard Analysis/Team Sheet/{team}.csv')
    all_players = team_sheet['Player Name'].tolist()
    teams_df[team] = all_players
```

```python
for team in teams:
    for index, row in df.loc[df['Team']==team].iterrows():
        current_name = row['Bowler'].split()
        for player in teams_df[team]:
            correct_name = player.split()
            if current_name == correct_name:
                pass
            elif current_name[-1] == correct_name[-1] and current_name[0] == correct_name[0][0]:
                current_name = correct_name
            else:
                pass
        if len(current_name) >= 2:
          current_name = f'{current_name[0]} {current_name[1]}'
          df.loc[index, 'Bowler'] = current_name
        else:
          df.loc[index, 'Bowler'] = current_name[0]


df
```

```python
df = df.drop(columns=['League'], axis=1)
df['3W'] = df['Wickets'].apply(lambda x: 1 if x == 3 else 0)
df['4W'] = df['Wickets'].apply(lambda x: 1 if x == 4 else 0)
df['5W'] = df['Wickets'].apply(lambda x: 1 if x >= 5 else 0)
df
```

```python
df.to_csv('/content/drive/MyDrive/Durham University/Dissertation/Dissertation/Scorecard Analysis v3/Bowling Scorecard Analysis/Bowling Scorecard (Correct Names).csv', index=False)
```

```python
def calculate_bowler_stats(group):
    total_matches = group.shape[0]
    total_overs_bowled = group['Overs'].sum()
    total_extras = group['Wides'].sum() + group['No Balls'].sum()
    total_balls_bowled = total_overs_bowled * 6 + total_extras
    abbm = total_balls_bowled / total_matches

    total_wickets = group['Wickets'].sum()
    total_runs_conceded = group['Runs'].sum()
    economy_rate = total_runs_conceded / total_overs_bowled
    strike_rate = total_wickets / (total_balls_bowled - total_extras) if total_wickets > 0 else 0
    cs = total_runs_conceded / total_wickets if total_wickets > 0 else 0

    total_3ws = group['3W'].sum()
    total_4ws = group['4W'].sum()
    total_5ws = group['5W'].sum()
    total_maidens = group['Maidens'].sum()

    total_matches_3w_plus = total_3ws + total_4ws + total_5ws
    total_wickets_3w_plus = (3 * total_3ws) + (4 * total_4ws) + (5 * total_5ws)
    total_wickets_less_3w = total_wickets - total_wickets_3w_plus

    return pd.Series({
        'Total Matches Played': total_matches,
        'Total Overs Bowled': total_overs_bowled,
        'Total Balls Bowled': total_balls_bowled,
        'Total Maidens': total_maidens,
        'Average Balls Bowled per Match (ABBM)': abbm,
        'WTS': strike_rate,
        'Total Wickets': total_wickets,
        'Total Runs Conceded': total_runs_conceded,
        'Economy Rate': economy_rate,
        'Total Extras': total_extras,
        'Total 3Ws': total_3ws,
        'Total 4Ws': total_4ws,
        'Total 5Ws': total_5ws,
        'Total Matches with 3W+': total_matches_3w_plus,
        'Total Wickets (3W+)': total_wickets_3w_plus,
        'Total Wickets (Less than 3W)': total_wickets_less_3w,
        'CS': cs
    })


df2 = df.groupby(['Bowler', 'Team']).apply(calculate_bowler_stats).reset_index()
df2
```

```python
import math
avg_overs = math.ceil(df2['Total Overs Bowled'].mean())
avg_overs
```

```python
df2 = df2[df2['Total Overs Bowled']>= avg_overs]
df2
```

```python
df2 = df2[df2['Total Matches Played'] >= 4]
df2
```

```python
df2 = df2.round(4)
```

```python
df2['ES'] = df2['Total Runs Conceded']/df2['Total Overs Bowled']

df2['WTS'] = df2['WTS']

df2['BBI'] = df2['Total Balls Bowled'] / df2['Total Matches Played']

df2['CS'] = df2['CS']

df2['BWTS'] = df2['Total Matches with 3W+'] / df2['Total Matches Played']

df2['SPS'] = (df2['Total Wickets']-df2['Total Wickets (3W+)']) / (df2['Total Matches Played'] - df2['Total Matches with 3W+'])

df2['MI'] = df2['Total Maidens'] / df2['Total Overs Bowled']

df2['EI'] = df2['Total Extras'] / df2['Total Matches Played']

df2['AVMS'] = (1/df2['ES']) * (df2['WTS']) * df2['Average Balls Bowled per Match (ABBM)']

columns_to_drop = [
    'Total Matches Played',
    'Total Overs Bowled',
    'Total Balls Bowled',
    'Total Maidens',
    'Average Balls Bowled per Match (ABBM)',
    'Total Wickets',
    'Total Runs Conceded',
    'Economy Rate',
    'Total Extras',
    'Total 3Ws',
    'Total 4Ws',
    'Total 5Ws',
    'Total Matches with 3W+',
    'Total Wickets (3W+)',
    'Total Wickets (Less than 3W)'
]

df2 = df2.drop(columns=columns_to_drop)

```
