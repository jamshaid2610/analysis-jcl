```python
all_teams = batting_scorecard['Batting Side'].unique().tolist()
```

```python
'''
The following code makes sures that all the players in the Batting Scorecard are in the Team sheet of a team. Some players are names in different
format. The code also fixes that and gives a consistent name to all the players. Detail of this in the project report.
'''
team_sheet = {'Player Name': None}
for team in all_teams:
    current_df = batting_scorecard[batting_scorecard['Batting Side']==team]
    players = current_df['Batsman'].unique().tolist()
    initials = []
    complete_names = []
    for name in players:
        current_name = name.split()
        if len(current_name[0])>1:
            complete_names.append(current_name)
        else:
            initials.append(current_name)
    new_players = []
    for i in initials:
        for j in complete_names:
            if i[-1]==j[-1] and i[0]==j[0][0]:
                i[0] = j[0]
            else:
                pass
        new_players.append(f'{i[0]} {i[-1]}')
    team_sheet['Player Name'] = new_players
    team_sheet_df = pd.DataFrame(team_sheet)
    team_sheet_df.to_csv(f'Team Sheet/{team}.csv', index=False)
```
